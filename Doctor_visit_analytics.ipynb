# Visualize and analyse the maximum,minimum and medium income

```python
y=list(df.income)
plt.boxplot(y)
plt.show()
```

# Find out the no:of days of reduced activity of male and female seperately due to illness

# Visualise is there any missing value in the dataset based on heat map

```python
#missing values
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

# Find out the correlation between variables in the goven dataset correlation between different variables

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='Blues')
```

# Analyse how the income of patient affects the no of visits to the hospital

```python
#correlation between income and visits
plt.figure(figsize=(10,10))
plt.scatter(x="income",y='visits',data=df)
plt.xlabel('income')
plt.ylabel('visits')
```

# count and visualize the no:of males and females affected by illness

```python
#no:of male and females affected by illness
sns.histplot(df.gender,bins=2)
```

# Visualize the percentage of people getting govt health insurance due to low income ,due to old age and also  the percentage of people having private insurance


```python
#% of people getting govt Insurance due to low income
label=['yes','no']
Y=df[df['freepoor']=='yes']
N=df[df['freepoor']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt health insurance due to low income")
plt.show()

#% of people having private Insurance
Y=df[df['private']=='yes']
N=df[df['private']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people having private Insurance")
plt.show()

#% of people getting govt Insurance due to old age,disability or veteran status
label=['yes','no']
Y=df[df['freerepat']=='yes']
N=df[df['freerepat']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt health insurance due to old age,disability or veteran status")
plt.show()
```

# Horizontal bar chart to analyse the reduced days of activity due to illness based on gender

```python
db=df.groupby('gender')['reduced'].sum().to_frame().reset_index()
#creating a bar chart
plt.barh(db['gender'],db['reduced'],color=['cornflowerblue','lightseagreen'])
#adding the aesthetics
plt.title("Bar Chart")
plt.xlabel("gender")
plt.ylabel('reduced activity')
#show plot
plt.show()
```
